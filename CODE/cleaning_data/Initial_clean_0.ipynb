```python
# string truncate and reshaping that applies to column 0 and 1
def truncate_name(str):
    out = ''
    #if it's empty then pass, apprently
    if str == "[]":
        #print('empty string')
        return 'no'
    else:
        for ch in str:
            #check if they are "[", "]", and "'" since they are not data
            if (ch != "[") and (ch != "]") and (ch != "'"): out += ch
        
    # done and return
    return out


# string truncate and reshaping that applies to column 2 to 5
def truncate_string(str):
    out = []
    temp = ''
    #if it's empty then pass, apprently
    if str == "[]":
        return 'no'
    else:
        for ch in str:
            #check if they are "[", "]", " ", and "'" since they are not data
            if (ch != "[") and (ch != "]") and (ch != ' ') and (ch != "'"):
                # if it is not ",", then it is data. We need to get the full string
                if (ch != ","):
                    temp += ch
                # otherwise append temp to the output string list and reset temp
                else:
                    out.append(temp)
                    temp = ''
 
        # one final time to do since there is no "," at the end
        out.append(temp)
        
    # done and return
    return out


# null killing and reshaping
def kill_null(str):
    out = str[0:2]
    temp = []
    for i in range(min(len(str[2]), len(str[3]), len(str[4]))):
        if (str[2][i] != 'null') and (str[3][i] != 'null') and (str[4][i] != 'null'):
            
            #add 
            if (str[2][i].find("\\") == -1 ) and (str[3][i].find("\\") == -1 ) and (str[4][i].find("\\") == -1 ):
                temp.append([str[2][i], str[3][i], str[4][i]])
            
    out.append(temp)
    return out
```

```python
# reshape and remove the [] rows
output_m = []
#for i in range(len(data)):
for i in range(len(data)):
    temp0 = truncate_name(data[i][0])
    # if temp0 is 'no', then the row is empty
    if temp0 != 'no':
        temp =[temp0, truncate_name(data[i][1]), truncate_string(data[i][2]), truncate_string(data[i][3]),truncate_string(data[i][4])]
        output_m.append(temp)
```

```python
# re-reshape and remove null data
output_n = []
for i in range(len(output_m)):
    print(i)
    output_n.append(kill_null(output_m[i]))
```

```python
# split every row
output_o = []
for i in range(len(output_n)):
    for j in range(len(output_n[i][2])):
        output_o.append([output_n[i][0], output_n[i][1], output_n[i][2][j][0], output_n[i][2][j][1], output_n[i][2][j][2]])
```
